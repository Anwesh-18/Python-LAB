## 1. Given a range of first 10 numbers, Iterate from start number to the end number and print the sum of the current number and previous number.

```python
a = 0
b = 1
print(a)
print(b)
for i in range(10):
    c = a+b
    a = b
    b = c
    print(c)
```

## 7. Given a two list of numbers create a new list such that new list should contain only odd numbers from the first list and even numbers from the second list.

```python
l1 = list(map(int,input().split()))
l2 = list(map(int,input().split()))
res = []
for num in l1:
    if num%2!=0:
        res.append(num)
for num in l2:
    if num%2==0:
        res.append(num)
print(res)
```

## 10. Find the sum of the series 2 +22 + 222 + 2222 + .. n terms.

## 11. Given a two list. Create a third list by picking an odd-index element from the first list and even index elements from second.

```python
l1 = list(map(int,input().split()))
l2 = list(map(int,input().split()))
res = []
i,j = 1,0
while i<len(l1) and j<len(l2):
    res.append(l1[i])
    res.append(l2[j])
    i+=2
    j+=2
    
print(res)
```

## 13. Given a list slice it into a 3 equal chunks and reverse each list.

```python
l = list(map(int,input().split()))
size = len(l)//3
res=[]
size_count=0
i=len(l)-1
while i>=0:
    if size_count >= size:
        print(res)
        res.clear()
        size_count=0
    else:
        res.append(l[i])
        size_count+=1
        i-=1
```

## 14. Given a two list of equal size create a set such that it shows the element from both lists in the pair.

## 16. Given 2 strings, s1 and s2, create a new string by appending s2 in the middle of s1.

```python
s1 = input("Enter the first string: ")
s2 = input("Enter the second string: ")
idx = len(s1)//2
x = list(s1)
x.insert(idx+1,s2)
print("".join(x))
```

## 17. Given 2 strings, s1, and s2 return a new string made of the first, middle and last character of each input string.

```python
s1 = input("Enter the first string: ")
s2 = input("Enter the second string: ")
mid1 = len(s1)//2
mid2 = len(s2)//2
new_str = s1[0]+s2[0]+s1[mid1]+s2[mid2]+s1[-1]+s2[-1]
print(new_str)
```
