## 1. Given a range of first 10 numbers, Iterate from start number to the end number and print the sum of the current number and previous number.

```python
a = 0
b = 1
print(a)
print(b)
for i in range(10):
    c = a+b
    a = b
    b = c
    print(c)
```

## 13. Given a list slice it into a 3 equal chunks and reverse each list.

```python
l = list(map(int,input().split()))
size = len(l)//3
res=[]
size_count=0
i=len(l)-1
while i>=0:
    if size_count >= size:
        print(res)
        res.clear()
        size_count=0
    else:
        res.append(l[i])
        size_count+=1
        i-=1
```

## 18. Arrange string characters such that lowercase letters should come first

```python
s = input("Enter the string: ")
lower = []
upper = []
for ch in s:
    if ch.islower():
        lower.append(ch)
    else:
        upper.append(ch)
print(''.join(lower + upper))
```

## 19. Count all lower case, upper case, digits, and special symbols from a given string.

```python
s = input("Enter the string: ")
lower = upper = digit = special = 0
for ch in s:
    if ch.islower():
        lower += 1
    elif ch.isupper():
        upper += 1
    elif ch.isdigit():
        digit += 1
    else:
        special += 1
print(f'lower:{lower}, upper:{upper}, digit: {digit}, special:{special}')
```

## 20. Given a string, return the sum and average of the digits that appear in the string, ignoring all other characters.

```python
s = input("Enter the string: ")
total = 0
count = 0
for ch in s:
    if ch.isdigit():
        total += int(ch)
        count += 1
average = total / count
if count == 0:
    print(f'total:{total},average:Not possible')
else:
    print(f'total:{total}, average:{average}')
```

## 22. Write a Python program to find the second largest number in a list.

```python
lst = [10, 20, 4, 45, 99]
largest = second = float('-inf')
for num in lst:
    if num > largest:
        second = largest
        largest = num
    elif num > second and num != largest:
        second = num
print("Second largest:", second)
```

## 23. Write a Python program to sort a list of elements using the bubble sort algorithm.

```python
lst = [64, 34, 25, 12, 22, 11, 90]
n = len(lst)
for i in range(n):
    for j in range(0, n - i - 1):
        if lst[j] > lst[j + 1]:
            lst[j],lst[j+1] = lst[j+1],lst[j]
print(lst)
```

## 25. Write a Python program to count the frequency of each element in a list.

```python
lst = [1, 2, 2, 3, 3, 3]
freq = {}
for i in lst:
    if i in freq:
        freq[i] += 1
    else:
        freq[i] = 1
print(freq)
```

## 30.Write a Python Program to print all Possible Combinations from the three Digits.

```python
digits = [1, 2, 3]
for i in digits:
    for j in digits:
        for k in digits:
            if i != j and j != k and i != k:
                print(i, j, k)
```
