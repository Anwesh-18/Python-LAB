## 1. Write a program to find the smallest and largest digit in a number.

```python
n = int(input("Enter number: "))
min_d = 9
max_d = 0
while n > 0:
    d = n % 10
    if d < min_d:
        min_d = d
    if d > max_d:
        max_d = d
    n //= 10
print("Smallest:", min_d)
print("Largest:", max_d)
```

## 2. Write a program to find the frequency of each digit in a number.

```python
n = int(input("Enter number: "))
for i in range(10):
    count = 0
    temp = n
    while temp > 0:
        if temp % 10 == i:
            count += 1
        temp //= 10
    if count > 0:
        print(i, ":", count)
```

## 3. Write a program to check whether a number is binary or not.

```python
n = int(input("Enter number: "))
found=True
while n > 0:
    d = n % 10
    if d != 0 and d != 1:
        found = False
        break
    n //= 10
if found:
    print("Binary number")
else:
    print("Not a binary number")
```

## 4.Write a program to count how many digits are even and odd in a number.

```python
n = int(input("Enter number: "))
even = 0
odd = 0
while n > 0:
    d = n % 10
    if d % 2 == 0:
        even += 1
    else:
        odd += 1
    n //= 10
print("Even digits:", even)
print("Odd digits:", odd)
```

## 6.Write a Python program to construct the following pattern, using a nested for loop.


```python
turn = 0
n = int(input("Enter number: "))
for i in range(1,n):
    if turn<(n//2):
        for j in range(i):
            print("*", end=" ")
        print()
    else:
        for j in range(n-i):
            print("*",end=" ")
        print()
    turn+=1
```

## 7. Print numbers from 1 to N, but skip numbers whose digit sum is divisible by 3.

```python
n = int(input("Enter N: "))
for i in range(1, n + 1):
    s = 0
    temp = i
    while temp > 0:
        s += temp % 10
        temp //= 10
    if s % 3 == 0:
        continue
    print(i,end=" ")
```

## 8.	Count numbers between 1 and 100 whose reverse is greater than the original number.

```python
count = 0
for i in range(1, 101):
    rev = 0
    temp = i
    while temp > 0:
        rev = rev * 10 + temp % 10
        temp //= 10
    if rev > i:
        count += 1
print(count)
```

## 9.	Print all numbers between 1 and 100 where product of digits > sum of digits.

```python
for i in range(1, 101):
    s = 0
    p = 1
    temp = i
    while temp > 0:
        d = temp % 10
        s += d
        p *= d
        temp //= 10
    if p > s:
        print(i,end=" ")
```

## 10.	Print all numbers from 1 to 100 that are not divisible by any digit they contain.

```python
for i in range(1, 101):
    temp = i
    flag = 1
    while temp > 0:
        d = temp % 10
        if d != 0 and i % d == 0:
            flag = 0
            break
        temp //= 10
    if flag == 1:
        print(i,end=" ")
```

## 12.	 Swap the first and last digit of a number.

```python
n = int(input("Enter number: "))
temp = n
last = n % 10
digits = 0
while temp > 0:
    digits += 1
    temp //= 10
first = n // (10 ** (digits - 1))
middle = (n % (10 ** (digits - 1))) // 10
new_num = last * (10 ** (digits - 1)) + middle * 10 + first
print(new_num)
```

## 14.	Check whether a number is a Harshad number (divisible by sum of digits).

```python
n = int(input("Enter number: "))
s = 0
temp = n
while temp > 0:
    s += temp % 10
    temp //= 10
if n % s == 0:
    print("Harshad number")
else:
    print("Not Harshad number")
```

## 19.	Find the LCM and GCD of two numbers.

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
x = a
y = b
while y != 0:
    x, y = y, x % y
gcd = x
lcm = (a * b) // gcd
print("GCD:", gcd)
print("LCM:", lcm)
```
