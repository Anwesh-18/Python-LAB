## 1. Write a Python program to count the number of lines, words, and characters in a file.

```python
f = open('file.txt')
lines = f.readlines()
word_count = 0
chars = 0
for line in lines:
    words = line.split()
    word_count += len(words)
    for word in words:
        chars += len(word)
f.close()
print(f'lines:{len(lines)}\nwords:{word_count}\ncharacters:{chars}')
```

## 7. Count uppercase letters, lowercase letters, and digits in a file.


```python
f = open('trial.txt')
lines = f.readlines()
f.close()
up_case = 0
low_case = 0
digit = 0
for line in lines:
    words = line.split()
    for word in words:
        for ch in word:
            if ch.isdigit():
                digit += 1
            elif ch.isupper():
                up_case += 1
            else:
                low_case += 1
print(f'upper case:{up_case}\nlower case:{low_case}\ndigits:{digit}')
```

## 8. Find the longest word in a file.


```python
f = open('file.txt')
lines = f.readlines()
res_len = float('-inf')
res = None
for line in lines:
    words = line.split()
    for word in words:
        if len(word) > res_len:
            res_len = len(word)
            res = word
f.close()
print(f'result: {res}')
```

## 13. Write a program to count number of vowels and consonants in a file


```python
f = open('trial.txt')
lines = f.readlines()
f.close()
vowels = ['a','e','i','o','u','A','E','I','O','U']
v = 0
c = 0
for line in lines:
    words = line.split()
    for word in words:
        for ch in word:
            if ch in vowels:
                v += 1
            else:
                c += 1
f.close()
print(f'vowels:{v}\nconsonents:{c}')
```

## 14. Write a program to reverse each word of a file and store in another file

```python
f1 = open('trial.txt')
f2 = open('new_trial.txt','w')
lines = f1.readlines()
f1.close()
for line in lines:
    rev = []
    words = line.split()
    for word in words:
        rev.append(word[::-1])
    new_line = " ".join(rev) + '\n'
    f2.write(new_line)
    
f2.close()
print('Operation completed')
```

## 19. Create the user exception class called InsufficientFundsError. Write a function withdraw(balance, amount) that raises this exception if the withdrawal amount exceeds the balance.

```python
class InsufficientFundsError(Exception):
    pass

def withdraw(balance, amount):
    if amount > balance:
        raise InsufficientFundsError(f"Insufficient funds: Tried to withdraw {amount}, but only {balance} available.")
    
    new_balance = balance - amount
    print(f"Withdrawal of {amount} successful. New balance: {new_balance}")
    return new_balance

try:
    current_balance = 500
    withdraw_amount = 600
    current_balance = withdraw(current_balance, withdraw_amount)
except InsufficientFundsError as e:
    print(f"Error: {e}")
```

## 21. Write a Python program to input student marks. If marks are less than 0 or greater than 100, raise a user-defined exception called InvalidMarksError.

```python
class InvalidMarksError(Exception):
    pass
subs = int(input("Total subjects: "))
marks = []
while len(marks)<subs:
    try:
        mark = int(input("Enter the mark: "))
        if mark < 0 or mark > 100:
            raise InvalidMarksError(f"Not a valid mark for a student, must be between 0-100!")
        
    except InvalidMarksError as i:
        print(f"Error:{i}")
    else:
        marks.append(mark)
        
print("Completed Successfully: ",marks)
```

## 27. ATM Withdrawal Limit: Write a Python program to simulate ATM withdrawal. If the withdrawal amount is greater than the account balance or exceeds Rs.20000 per day, raise a user-defined exception.

```python
class InsufficientFundsError(Exception):
    pass

class Bank:
    def __init__(self,bal,tw):
        self.balance = bal
        self.tod_withdrawal = tw
    
    def withdraw(self, amount):
        if amount > self.balance or self.tod_withdrawal>20000:
            raise InsufficientFundsError(f"Insufficient funds: Tried to withdraw {amount}")
    
        self.balance = self.balance - amount
        self.tod_withdrawal += amount
        print(f"Withdrawal of {amount} successful. New balance: {self.balance}")
    
b = Bank(25000,21000)
try:
    bal = b.withdraw(5010)
except InsufficientFundsError as e:
    print(f"Error: {e}")
```

## 29. Product Stock Management: Write a Python program for a product inventory system If the user tries to purchase more items than available in stock, raise a user-defined exception called OutOfStockError.

```python
class OutOfStockError(Exception):
    pass

class Stock:
    def __init__(self,avail):
        self.avail = avail
    
    def purchase(self, amount):
        if amount > self.avail:
            raise OutOfStockError(f"Insufficient stocks: Stock Left:{self.avail}")
    
        self.avail -= amount
        print(f"{amount} amount of stock purchased successfully")
    
b = Stock(25)
try:
    b.purchase(30)
except OutOfStockError as e:
    print(f"Error: {e}")
```
