### 2. Rectangle Area Create a class Rectangle with: length ,width And Add methods: area() → returns area ,perimeter() → returns perimeter

```python
class Rectangle:
    def __init__(self,length,width):
        self.length = length
        self.width = width
        
    def area(self):
        return self.length*self.width
    def perimeter(self):
        return 2*self.length*self.width

r = Rectangle(20,40)
print(r.area())
print(r.perimeter())
```

### 4. Create a class Circle with attribute radius. Add methods: • area(), • circumference() Use the value π = 3.14.

```python
class Circle:
    def __init__(self,rad):
        self.rad = rad
    
    def area(self):
        return (3.14)*((self.rad)**2)
    
    def circumference(self):
        return 2*(3.14)*self.rad
    
c = Circle(12)
print(c.area())
print(c.circumference())
```

### 8.Create a class Temperature with attribute celsius. Methods: • to_fahrenheit() • to_kelvin()

```python
class Temperature:
    def __init__(self,celsius):
        self.celsius = celsius
    
    def to_fahrenheit(self):
        return self.celsius*(9/5) + 32
    
    def to_kelvin(self):
        return self.celsius+273
    
t = Temperature(40)
print(t.to_fahrenheit())
print(t.to_kelvin())
```

### 10. Create a class Cart with: • items (dictionary)
### Methods: • add_item(item, price) • remove_item(item) • total_price()

```python
class Cart:
    def __init__(self,items: dict):
        self.items = items
        
    def add_item(self,item,price):
        self.items[item] = price
    
    def remove_item(self,item):
        del self.items[item]
    
    def total_price(self):
        total = 0
        for val in self.items.values():
            total += val
        return total
    
c = Cart({"apple":20,"banana":10})
c.add_item("mango",30)
c.remove_item("apple")
print(c.total_price())
```

### 11. Create a class Student with: • name • marks (list)
### Methods: • average_marks() • grade() • high_grad() 
### Grades: • A → 90+ • B → 75+ • C → 50+ • F → below 50

```python
class Student:
    def __init__(self,name,marks):
        self.name = name
        self.marks = marks
    
    def average_marks(self):
        return sum(self.marks)/len(self.marks)
    
    def grade(self,mark):
        if mark > 90:
            return "A"
        elif mark > 75:
            return "B"
        elif mark >= 50:
            return "C"
        else:
            return "F"
        
    def high_grad(self):
        max_g = ""
        max_m = float('-inf')
        for mark in self.marks:
            if mark > max_m:
                max_m = mark
                max_g = self.grade(mark)
        
        return max_g

s = Student("Anwesh",[90,96,95,94,98])
print(s.average_marks())
print(s.high_grad())
```

## 14. Create a Person class with attributes like name and date_of_birth. Implement a method that calculates the person's current age based on the date of birth.

```python
from datetime import date
class Person:
    def __init__(self,name,dob):
        self.name = name
        self.dob = dob.split('-')
        
    def calculate_age(self):
        curr = str(date.today()).split('-')
        years = int(curr[0])-int(self.dob[0])
        months = int(curr[1])-int(self.dob[1])
        if months<0:
            years-=1
            months %= 12
        return f"{years} Years {months} months\nAge: {years}"

p = Person('Anwesh','2006-06-26')
print(p.calculate_age())
```

## 16. Define a class Point, which can be used to create two-dimensional points. Create three objects of Point class and check, these points are non collinear.

```python
class Point:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    
    def check_coll(self,p2,p3):
        if self.x//self.y == p2.x//p2.y == p3.x//p3.y:
            return True
        else:
            return False

p1 = Point(1,2)
p2 = Point(2,3)
p3 = Point(4,5)
print(p1.check_coll(p2,p3))
```

## 18. Define a class ComplexNumber, placed the necessary data members and member functions to perform basic Complex number operation.

```python
class Complex:
    def __init__(self,real,img):
        self.real = real
        self.img = img
        
    def __add__(self,c):
        real = self.real + c.real
        img = self.img + c.img
        if img < 0:
            return f"{real} - {abs(img)}i"
        else:
            return f"{real} + {img}i"
        
    def __sub__(self,c):
        real = self.real - c.real
        img = self.img - c.img
        if img < 0:
            return f"{real} - {abs(img)}i"
        else:
            return f"{real} + {img}i"
c1 = Complex(2,3)
c2 = Complex(-3,-9)
print(c1+c2)
print(c1-c2)
```

## 19. Define a class Time having data members hour, minute and second. The class should have a parameterized constructor to initialize its data members. It should also have methods displayTime() to display the time in HH:MM: SS format and addTime() to add two-time objects.

```python
class Time:
    def __init__(self,hours,minutes,seconds):
        self.hours = hours
        self.minutes = minutes
        self.seconds = seconds
    def display_time(self):
        return f"{self.hours}:{self.minutes}:{self.seconds}"
            
    def __add__(self,t):
        hour = self.hours + t.hours
        minute = self.minutes + t.minutes
        if minute >= 60:
            hour += 1
            minute %= 60
        second = self.seconds + t.seconds
        if second >= 60:
            minute += 1
            second %= 60
        return f"{hour}:{minute}:{second}"
t1 = Time(2,30,45)
t2 = Time(3,45,20)
print(t1+t2)
```

## 23. Develop a Python program that will deal with employee information of an organization. Define a class Employee, which has the data members and member functions are follows.
## Data Mambers - empName - empNumber - basicSal - DA - HRA
## Member Functions - calGrossSal ( ) - showEmpDetails( )
## Create an array of objects. Display those employees, whose gross salary is more than 50,000. Note: You are free to add more number of relevant data member and member function.

```python
class Employee:
    def __init__(self,empName,empNumber,basic_sal,DA,HRA):
        self.empName = empName
        self.empNumber = empNumber
        self.basic_sal = basic_sal
        self.DA = DA
        self.HRA = HRA
    def calGrossSal(self):
        return self.basic_sal + self.DA + self.HRA
        
    def showEmpDetails(self):
        print(f"Name: {self.empName}\nemp_number: {self.empNumber}\nGross_salary: {self.basic_sal + self.DA + self.HRA}")

emp_list = []
for i in range(3):
    name = input("Enter the name: ")
    num = int(input("Enter the empnumber: "))
    basic_salary = int(input("Enter the basic salary: "))
    da = int(input("Enter DA: "))
    hra = int(input("Enter HRA: "))
    e = Employee(name,num,basic_salary,da,hra)
    emp_list.append((e,e.calGrossSal()))
for e in emp_list:
    if e[1] > 50000:
        e[0].showEmpDetails()
```
