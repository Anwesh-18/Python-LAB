## 4. Find the number of occurrences for any repeated item present in a tuple.

```python
freq = {}
tup = (1,1,2,1,'Hello',23.67)
for t in tup:
    freq[t] = freq.get(t,0)+1
for key,val in freq.items():
    if val >= 2:
        print(key)
```

## 5. Display the elements present in a tuple which are at the even position.

```python
tup = (True,1,2,1,'Hello',23.67)
for i in range(len(tup)):
    if i%2 == 0:
        print(tup[i])
```

## 6. Remove the elements of a tuple which are present in the odd index position.

```python
tup = (True,1,2,1,'Hello',23.67)
for i in range(len(tup)):
    if i%2 != 0:
        print(tup[i])
```

## 7. Remove the odd number elements from a tuple.

```python
tup = (True,1,2,1,'Hello',23.67)
l = list(tup)
for i in range(len(tup)):
    if i%2 != 0:
        l.remove(tup[i])
print(tuple(l))
```

## 14. Display the largest number of a tuple without using the built-in function max().The user need to input the values in tuple from the keyboard.

```python
tup = tuple(map(int,input("Enter the values: ").split()))
largest = float('-inf')
for t in tup:
    if t > largest:
        largest = t
        
print(largest)
```

## 15. Replace the last value of the tuple with 100.Sample list: [(1, 2, 3), (4, 5, 6), (7, 8, 9)] Output: [(1, 2, 100), (4, 5, 100),(7, 8, 100)]

```python
l = [(1, 2, 3), (4, 5, 6), (7, 8, 9)]
res = []
for t in l:
    temp = list(t)
    temp[-1] = 100
    res.append(tuple(temp))
print(res)
```

## 16. Sort a tuple by its float element. Sample data: [('item1', '12.20'), ('item2','15.10'), ('item3', '24.5')] Output: [('item3', '24.5'), ('item2', '15.10'),('item1', '12.20')]

```python
l = [('item1', '12.20'), ('item2','15.10'), ('item3', '24.5')]
l.sort(key=lambda x: float(x[1]),reverse=True)
print(l)
```

## 17. Count the number of elements in a tuple until an element is a list.

```python
tup = (True,False,12,34,[1,2,3],43,65)
count = 0
for t in tup:
    if isinstance(t,list):
        break
    count += 1
print(count)
```
