## 16.Find the roots of a quadratic equation.


```python
import math
a = 1
b = -3
c = 2
d = b*b - 4*a*c
root1 = (-b + math.sqrt(d)) / (2*a)
root2 = (-b - math.sqrt(d)) / (2*a)
print(root1, root2)
```

## 17.Accept basic salary from user. Calculate DA(50%), HRA(20%)  and find the gross salary.(input:20000)

```python
basic = float(input("Enter basic salary: "))
DA = basic * 0.50
HRA = basic * 0.20
gross = basic + DA + HRA
print(gross)
```
