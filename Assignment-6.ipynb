## Q2. Write a program to convert a number entered by the user into its corresponding number in words. For example, if the input is 876 then the output should be 'Eight Seven Six'. (Hint. use dictionary for keys 0-9 and their values as equivalent words.)

```python
vals = {1:'One ',2:'Two ',3:'Three ',4:'Four ',5:'Five ',6:'Six ',7:'Seven ',8:'Eight ',9:'Nine ',0:'Zero '}
num = int(input("Enter the number: "))
res = []
while num >0:
    digit = num%10
    res.append(vals[digit])
    num //=10
new_res = res[::-1]
print("".join(new_res))
```

## Q3. Create a dictionary whose keys are month names and whose values are the number of days in the corresponding months. (a) Ask the user to enter a month name and use the dictionary to tell how many days are in the month. (b) Print out all of the keys in alphabetical order. (c) Print out all of the months with 31 days. (d) Print out the (key-value) pairs sorted by the number of days in each month.

```python
d = {'january':31,'february':28,'march':31,'april':30,'may':31,'june':30,'july':31,'august':31,'september':30,'october':31,'november':30,'december':31}
month = input("Enter the month name: ")
print(f'{month} has {d[month]} number of Days!!')

# Print out all of the keys in alphabetical order.
keys = []
for key,val in d.items():
    keys.append(key)
keys.sort()
print(keys)

# Print out all of the months with 31 days.
for key,val in d.items():
    if val == 31:
        print(key)

# Print out the (key-value) pairs sorted by the number of days in each month.
sorted_by_days = sorted(d,key=lambda x: d[x])
print(sorted_by_days)
```

## Q4. Given the dictionary x = {'k1':'v1', 'k2':'v2', 'k3':'v3'}, create a dictionary with the opposite mapping, i.e., write a program to create the dictionary as : inverted_x = {'v1': 'k1' , 'v2' :'k2' , 'v3':'k3'}

```python
x = {'k1':'v1', 'k2':'v2', 'k3':'v3'}
inverted_x = {}

for key,val in x.items():
    if val in inverted_x:
        inverted_x[val].append(key)
    else:
        inverted_x[val] = [key]
print(inverted_x)
```

## Q5. Given two dictionaries say D1 and D2. Write a program that lists the overlapping keys of the two dictionaries, i.e., if a key of D1 is also a key of D2, then list it.

```python
d1 = {1:'one',2:'two',3:'three',4:'four',5:'five'}
d2 = {4:'four',5:'five',6:'six',7:'seven',8:'eight'}
res = []
for key,val in d1.items():
    if key in d2:
        res.append(key)
print(res if res else "No common keys!")
```

## Q6. Write a program that checks if two same values in a dictionary have different keys. That is, for dictionary D1 = { 'a' : 10, 'b': 20, 'c' : 10}, the program should print 2 keys have same values and for dictionary D2 = {'a' : 10, 'b' : 20, 'c' : 30}, the program should print No keys have same values.

```python
d = {'a' : 10, 'b' : 20, 'c' : 30}
seen = set()
count = 1
for key,val in d.items():
    if val not in seen:
        seen.add(val)
    else:
        count += 1
print(f'{count} keys have same values' if count>1 else "No keys have same values")
```

## Q7. Write a program to check if a dictionary is contained in another dictionary e.g., if d1 = {1:11, 2:12} d2 = {1:11, 2:12, 3:13, 4:15} then d1 is contained in d2.

```python
d1 = {1:11, 2:12}
d2 = {1:11, 3:13, 4:15}

length = 0
dsize = 0
for key,val in d1.items():
    dsize += 1
    if key in d2 and val == d2[key]:
        length += 1
print("d1 is contained in d2" if length==dsize else "d1 is not contained in d2")
```

## Q8. A dictionary D1 has values in the form of lists of numbers. Write a program to create a new dictionary D2 having same keys as D1 but values as the sum of the list elements e.g., D1 = {'A' : [1, 2, 3] , 'B' : [4, 5, 6]} then D2 is {'A' :6, 'B' : 15}

```python
d = {'A' : [1, 2, 3] , 'B' : [4, 5, 6]}
new_d = {}
for key,val in d.items():
    new_d[key] = sum(val)
print(new_d)
```

## Q9. Create a dictionary ‘ODD’ of odd numbers between 1 and 10, where the key is the decimal number and the value is the corresponding number in words. Perform the following operations on this dictionary: 
## (a) Display the keys
## (b) Display the values
## (c) Display the items
## (d) Find the length of the dictionary
## (e) Check if 7 is present or not
## (f) Check if 2 is present or not
## (g) Retrieve the value corresponding to the key 9
## (h) Delete the item from the dictionary corresponding to the key 9

```python
odd = {1:'one',3:'three',5:'five',7:'seven',9:'nine'}
is_2 = False
is_7 = False
# keys
for key in odd.keys():
    if key == 2:
        is_2 = True
    if key == 7:
        is_7 = True
    print(key)
# vals
for key in odd.values():
    print(key)
length = 0
for item in odd.items():
    length += 1
    print(item)
print("length: ",length)
print("is 2 present?:",is_2)
print("is 7 present?:",is_7)
print("value of 9:",odd[9])
del odd[9]
print("is 9 present?:",odd.get(9))
```

## Q11. Word Frequency in a Paragraph

```python
par = "Paragraphs are the building blocks of papers. Many students define paragraphs in terms of length: a paragraph is a group of at least five sentences, a paragraph is half a page long, etc. In reality, though, the unity and coherence of ideas among sentences is what constitutes a paragraph. A paragraph is defined as “a group of sentences or a single sentence that forms a unit” (Lunsford and Connors 116). Length and appearance do not determine whether a section in a paper is a paragraph. For instance, in some styles of writing, particularly journalistic styles, a paragraph can be just one sentence long. Ultimately, a paragraph is a sentence or group of sentences that support one main idea. In this handout, we will refer to this as the “controlling idea,” because it controls what happens in the rest of the paragraph."
freq = {}
words = par.split()
for word in words:
    freq[word] = freq.get(word,0)+1
print(freq)
```

## Q12. Find Common Keys in Two Dictionaries

```python
d1 = {1:11, 2:12,3:34}
d2 = {1:11, 3:13, 4:15}
res = []
for key1 in d1.keys():
    for key2 in d2.keys():
        if key1==key2:
            res.append(key2)
print(res)
```

## Q13. Invert Dictionary with Duplicate Value
## Input: d = {'a':1, 'b':2, 'c':1, 'd':3, 'e':2}
## Output: {1: ['a', 'c'], 2: ['b', 'e'], 3: ['d']}

```python
x = {'a':1, 'b':2, 'c':1,'d':3, 'e':2}
inverted_x = {}

for key,val in x.items():
    if val in inverted_x:
        inverted_x[val].append(key)
    else:
        inverted_x[val] = [key]
print(inverted_x)
```

## Q16. Create a dictionary where keys are word lengths and values are lists of words.
## Input: ["cat","dog","elephant","rat","tiger"] Output: {3:['cat','dog','rat'], 5:['tiger'],8:['elephant']}

```python
inp = list(map(str,input("Enter the list:").split()))
d = {}
for el in inp:
    if len(el) in d:
        d[len(el)].append(el)
    else:
        d[len(el)] = [el]
print(d)
```
