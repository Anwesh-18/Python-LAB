## Question 1

```python
units = int(input("Enter the number: "))
total_cost = 0
if units>50 and units<=150:
    total_cost += (50*(0.5)) + ((units-50)*(0.75))
elif units>150 and units<=250:
    total_cost += (50*(0.5)) + (100*(0.75)) + ((units-150)*(1.2))
elif units>250:
    total_cost += (50*(0.5)) + (100*(0.75)) + ((100)*(1.2)) + ((units-250)*(1.5))
else:
    total_cost += units*0.5
print(total_cost)
```

## Question 8

```python
phy = int(input("Enter the marks of phy: "))
chem = int(input("Enter the marks of chem: "))
bio = int(input("Enter the marks of Bio: "))
math = int(input("Enter the marks of Math: "))
comp = int(input("Enter the marks of Comp: "))

total = phy+chem+bio+math+comp
per = (total/500)*100

if per>=90:
    print("Grade: A")
elif per >=80 and per <90:
    print("Grade; B")
elif per >=70 and per <80:
    print("Grade; C")
elif per >=60 and per <70:
    print("Grade; D")
elif per >=40 and per<60:
    print("Grade: E")
else:
    print("FAIL")
```

## Question 9

```python
side1 = int(input("Enter the first side: "))
side2 = int(input("Enter the second side: "))
side3 = int(input("Enter the third side: "))

if side1+side2 > side3 or side1+side3 >side2 or side2+side3 > side1:
    if side1 == side2 == side3:
        print("Equilateral")
    elif side1 == side2 or side2==side3 or side1==side3:
        print("isosceles")
    else:
        print("scalen")
else:
    print("Not a valid Triangle!!")
```
